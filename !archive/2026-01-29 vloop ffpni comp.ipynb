```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# basis conversion testing
import xarray as xr
filepath = 'tmp/test.nc'
data_tree = xr.open_datatree(filepath).sel(time=100.0, method='nearest')

psi_norm = data_tree.profiles.psi_norm.sel(method='nearest').to_numpy() # j_ohmic in A/m^2 on rho_N grid
rho_face_norm = data_tree.profiles.rho_face_norm.sel(method='nearest').to_numpy() # rho_N at cell faces

j_ohmic = data_tree.profiles.j_ohmic.sel(method='nearest').to_numpy() # j_ohmic in A/m^2 on rho_N grid
rho_cell_norm = data_tree.profiles.rho_cell_norm.sel(method='nearest').to_numpy() # rho_N at cell centers

plt.plot(rho_face_norm, psi_norm)

print(f'rho_cell_norm = {rho_cell_norm}')
print(f'rho_face_norm = {rho_face_norm}')
```
